```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import ipywidgets as widgets
from ipywidgets import interact
```

Read in files

```python
dir = '/content/drive/MyDrive/ServiceProjects-public/CivilRightCenter-RJC/Data Cleanup team/HMDA CSVs'
#array of data frames
df = []

def readInFile(file):
  df.append(pd.read_csv(file))


for file in os.listdir(dir):
  if file.endswith(".csv"):
    readInFile(dir + '/' + file)
    print(file)
```

Combine dataframes

```python
concatenated_df = pd.concat(df, ignore_index=True)
```

#Vizualizations

Group by income and loan type

```python
#plotly
def plot_data(year):
    # Filter the DataFrame for the selected year
    filtered_data = selected_columns[selected_columns['as_of_year'] == year]

    # Group and count the filtered data
    bar_data = filtered_data.groupby(['applicant_race_name_1', 'action_taken_name']).size().unstack(fill_value=0)

    # Plotting
    plt.figure(figsize=(8, 5))
    bar_data.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#4CAF50', '#FFC107', '#2196F3'])
    plt.title(f'Total Value by Category for Year {year}')
    plt.ylabel('Total Value')
    plt.xlabel('Category')
    plt.xticks(rotation=80)
    plt.tight_layout()  # Adjust layout to fit the labels
    plt.show()

#year_slider = widgets.IntSlider(
    #value=2007,  # Default starting year
    #min=selected_columns['as_of_year'].min(),  # Minimum year from the DataFrame
    #max=selected_columns['as_of_year'].max(),  # Maximum year from the DataFrame
    #step=1,
    #description='Select Year:',
    #continuous_update=False  # Only update when the slider is released
#)

#widgets.interactive(plot_data, year=year_slider)

# Display the slider
#display(year_slider)
interact(plot_data, year=(2007, 2017))
```
